# A simple script to rename real images and mask

```python
import os
import shutil
```

```python
real_image_dir = "/home/vajira/DL/temp_data_2/real_images"
real_image_dir_new = "/home/vajira/DL/temp_data_2/real_images_new"

real_mask_dir = "/home/vajira/DL/temp_data_2/real_masks"
real_mask_dir_new = "/home/vajira/DL/temp_data_2/real_masks_new"
```

```python
os.makedirs(real_image_dir_new, exist_ok=True)
os.makedirs(real_mask_dir_new, exist_ok=True)
```

```python
def move_and_add_one_to_name(src_dir, dst_dir):
    
    for img in os.listdir(src_dir):
        print(img)
        img_id = int(img.split(".")[0])
        new_img_name = str(img_id + 1) + "." +  img.split(".")[-1]
        shutil.copy(os.path.join(src_dir, img), os.path.join(dst_dir, new_img_name))
        
```

```python
move_and_add_one_to_name(real_mask_dir, real_mask_dir_new)
```
